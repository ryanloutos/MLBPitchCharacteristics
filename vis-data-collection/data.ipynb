```python
for index, row in data.iterrows():
    if row['pitch_type'] in pitcher_data[row['pitcher']]['pitches'].keys():
        pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['num'] += 1
        pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['velo_total'] += row['release_speed']
        pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['spin_rate_total'] += row['release_spin_rate']
        pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['pfx_z_total'] += row['pfx_z']
        pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['pfx_x_total'] += row['pfx_x']
        # pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['plate_x'].append(round(row['plate_x'], 2))
        # pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['plate_z'].append(round(row['plate_z'], 2))
        if row['description'] == 'swinging_strike':
            pitcher_data[row['pitcher']]['pitches'][row['pitch_type']]['swing_and_miss_total'] += 1

    # pitch counts
    pitch_count = f'{int(row["balls"])}-{int(row["strikes"])}'
    if pitch_count in pitcher_data[row['pitcher']]['pitch_counts'].keys():
        if row['pitch_type'] in pitcher_data[row['pitcher']]['pitch_counts'][pitch_count].keys():
            pitcher_data[row['pitcher']]['pitch_counts'][pitch_count][row['pitch_type']] += 1
            pitcher_data[row['pitcher']]['pitch_counts'][pitch_count]['num'] += 1
            if row['description'] == 'ball':
                pitcher_data[row['pitcher']]['pitch_counts'][pitch_count]['ball'] += 1
            else:
                pitcher_data[row['pitcher']]['pitch_counts'][pitch_count]['strike'] += 1

```

```python
json_data = []
for pitcher in pitcher_data.keys():
    this_data = {
        'name': pitcher_data[pitcher]['name'],
        'RHP': pitcher_data[pitcher]['RHP'],
        'pitches': {},
        'pitch_counts': {}
    }
    total_pitches = 0
    for pitch in ['FF', 'SI', 'FC', 'CH', 'SL', 'CU', 'KC', 'FS']:
        total_pitches += pitcher_data[pitcher]['pitches'][pitch]['num']
        if pitcher_data[pitcher]['pitches'][pitch]['num'] != 0:
            this_data['pitches'][pitch] = {
                'velocity': round(pitcher_data[pitcher]['pitches'][pitch]['velo_total'] / pitcher_data[pitcher]['pitches'][pitch]['num'], 1),
                'spin_rate': round(pitcher_data[pitcher]['pitches'][pitch]['spin_rate_total'] / pitcher_data[pitcher]['pitches'][pitch]['num'], 0),
                'vert_movement': round(pitcher_data[pitcher]['pitches'][pitch]['pfx_z_total'] / pitcher_data[pitcher]['pitches'][pitch]['num'], 2),
                'hor_movement': round(pitcher_data[pitcher]['pitches'][pitch]['pfx_x_total'] / pitcher_data[pitcher]['pitches'][pitch]['num'], 2),
                'swing_miss': round(100 * pitcher_data[pitcher]['pitches'][pitch]['swing_and_miss_total'] / pitcher_data[pitcher]['pitches'][pitch]['num']),
                # 'plate_x':  pitcher_data[pitcher]['pitches'][pitch]['plate_x'],
                # 'plate_y':  pitcher_data[pitcher]['pitches'][pitch]['plate_z']
            }
        else:
            this_data['pitches'][pitch] = {
                'velocity': 0,
                'spin_rate': 0,
                'vert_movement': 0,
                'hor_movement': 0,
                'swing_miss': 0,
                # 'plate_x':  0,
                # 'plate_y':  0,
            }
    for pitch in ['FF', 'SI', 'FC', 'CH', 'SL', 'CU', 'KC', 'FS']:
        this_data['pitches'][pitch]['usage'] = round(100 * pitcher_data[pitcher]['pitches'][pitch]['num'] / total_pitches, 1)
        if this_data['pitches'][pitch]['velocity'] != 0:
            this_data['pitches'][pitch]['bauer_units'] = round(this_data['pitches'][pitch]['spin_rate'] / this_data['pitches'][pitch]['velocity'], 1)
        else:
            this_data['pitches'][pitch]['bauer_units'] = 0

    for count in ['0-0', '0-1', '0-2', '1-0', '1-1', '1-2', '2-0', '2-1', '2-2', '3-0', '3-1', '3-2']:
        this_data['pitch_counts'][count] = {}
        for pitch in ['FF', 'SI', 'FC', 'CH', 'SL', 'CU', 'KC', 'FS']:
            if pitcher_data[pitcher]['pitch_counts'][count]['num'] != 0:
                this_data['pitch_counts'][count][pitch] = round(100 * pitcher_data[pitcher]['pitch_counts'][count][pitch] / pitcher_data[pitcher]['pitch_counts'][count]['num'], 1)
            else:
                this_data['pitch_counts'][count][pitch] = 0
        this_data['pitch_counts'][count]['num'] = pitcher_data[pitcher]['pitch_counts'][count]['num']
        if pitcher_data[pitcher]['pitch_counts'][count]['num'] != 0:
            this_data['pitch_counts'][count]['ball_pct'] =  round(100 * pitcher_data[pitcher]['pitch_counts'][count]['ball'] / pitcher_data[pitcher]['pitch_counts'][count]['num'], 1)
            this_data['pitch_counts'][count]['strike_pct'] =  round(100 * pitcher_data[pitcher]['pitch_counts'][count]['strike'] / pitcher_data[pitcher]['pitch_counts'][count]['num'], 1)
        else:
            this_data['pitch_counts'][count]['ball_pct'] =  0
            this_data['pitch_counts'][count]['strike_pct'] =  0
    
    json_data.append(this_data)
```

```python
pitcher_fb_json = []
for pitcher in json_data:
    this_data = {
        'name': pitcher['name'],
        'RHP': pitcher['RHP'],
        'data': {}
    }
    FF = pitcher['pitches']['FF']['usage']
    SI = pitcher['pitches']['SI']['usage']
    FC = pitcher['pitches']['FC']['usage']
    if FF >= SI and FF >= FC:
        this_data['data'] = pitcher['pitches']['FF']
    elif SI >= FF and SI >= FC:
        this_data['data'] = pitcher['pitches']['SI']
    else:
        this_data['data'] = pitcher['pitches']['FC']
    pitcher_fb_json.append(this_data)
```
